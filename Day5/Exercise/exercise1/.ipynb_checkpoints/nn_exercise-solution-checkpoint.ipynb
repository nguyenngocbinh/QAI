Let's build a 4-layer neural network. Our network has one input layer, two hidden layer and one output layer. Our model can be represented as a directed acyclic graph wherein each node in a layer is connected all other nodes in its succesive layer. The neural net is shown below:

![alt text](images/nn.svg "neural network")

**Firstly, we need to initialize some hyperparameters**

```python
input_dim = 2
hidden_dim = 8
output_dim = 1
```

### How our network makes predictions

Our network makes predictions using *forward propagation*, which is just a bunch of matrix multiplications and the application of the activation function(s) we defined above. If $x$ is the 2-dimensional input to our network then we calculate our prediction $\hat{y}$ (also two-dimensional) as follows:

$$
\begin{aligned}
z_1 & = xW_1 + b_1 \\
a_1 & = \tanh(z_1) \\
z_2 & = a_1W_2 + b_2 \\
a_2 & = \hat{y} = \mathrm{softmax}(z_2)
\end{aligned}
$$

$z_i$ is the weighted sum of inputs of layer $i$ (bias included) and $a_i$ is the output of layer $i$ after applying the activation function. $W_1, b_1, W_2, b_2$ are  parameters of our network, which we need to learn from our training data. You can think of them as matrices transforming data between layers of the network. Looking at the matrix multiplications above we can figure out the dimensionality of these matrices. If we use 500 nodes for our hidden layer then $W_1 \in \mathbb{R}^{2\times500}$, $b_1 \in \mathbb{R}^{500}$, $W_2 \in \mathbb{R}^{500\times2}$, $b_2 \in \mathbb{R}^{2}$. Now you see why we have more parameters if we increase the size of the hidden layer.

### Activation function

I will demonstrate some common activation fu

**Sigmoid function**

**Forward Pass**

$$out = \frac{1}{1 + e^{-scores}}$$


```python
scores_p = np.arange(-10, 10, 0.1)
out_p = 1. / (1. + np.exp(-scores_p))
plt.plot(scores_p, out_p)
plt.xlabel('in')
plt.ylabel('out')
```

```python
def sigmoid(input):
    #scores: N x K
    out = 1.0 / (1.0 + np.exp(-input))
    return out

def derivative_sigmoid(input):
    return sigmoid(input) * (1 - sigmoid(input))
```

**Tanh**

**Forward Pass**

$$out = \tanh(scores)$$


```python
scores_p = np.arange(-10, 10, 0.1)
out_p = np.tanh(scores_p)
plt.plot(scores_p, out_p)
plt.xlabel('in')
plt.ylabel('out')
```

```python
def tanh(input):
    #scores: N x K
    out = np.tanh(input)
    return out

def derivative_tanh(input):
    return 1 - tanh(input) ** 2
```

**ReLU**

**Forward Pass**

$$out = max(0,scores)$$

```python
scores_p = np.arange(-10, 10, 0.1)
out_p = np.maximum(0,scores_p)
plt.plot(scores_p, out_p)
plt.xlabel('in')
plt.ylabel('out')
```

```python
def relu(self,scores):
    #scores: N x K
    out = scores * (scores > 0)
    return out
def derivative_relu(input):
    return 1.0 * (input > 0)
```

### Define the model

**Now, we will define input, hidden and output layer**

```python
# layer 1 === input---> hidden === (W1,b1)
W1 = np.random.randn(input_dim, hidden_dim) # initialize weight value
b1 = np.zeros((1, hidden_dim)) # initialize bias value

# layer 2 === hidden---> hidden === (W2,b2)
W2 = np.random.randn(hidden_dim, hidden_dim)
b2 = np.zeros((1, hidden_dim))

# layer 3 === hidden---> output === (W3,b3)
W3 = np.random.randn(hidden_dim, output_dim)
b3 = np.zeros((1, output_dim))
```

**Feedforward function**

![alt text](images/feedforward.png "Feedforward")

```python
def feedforward(X):
    # Forward prop
    input = X
    pre_act1 = X.dot(W1) + b1 # input layer -> hidden layer 1
    act1 = tanh(pre_act1) # activation function of hiddin layer 1
    pre_act2 = act1.dot(W2) + b2 # # hidden layer 1 -> hidden layer 2
    act2 = tanh(pre_act2) # activation function of hidden layer 2
    pre_act3 = act2.dot(W3) + b3 # hidden layer 2 -> ouput layer
    act3 = sigmoid(pre_act3) # activation function of output layer
    return pre_act1, pre_act2, pre_act3, act1, act2, act3

dummyX = np.random.randn(10, 2) # 10 is number of data point
pre_act1, pre_act2, pre_act3, act1, act2, act3 = feedforward(dummyX)

print(f"Shape of hidden layer 1 (before activation function): {pre_act1.shape}")
print(f"Shape of hidden layer 2 (before activation function): {pre_act2.shape}")
print(f"Shape of output layer (before activation function): {pre_act3.shape}")
```

**Loss Function**

After caculate feedforward pass, we need to caculate error value between predicted output and real value.

There are many type of loss function such as MSE (mean square error), MAE (mean absolute error), Cross-entropy losss, ... In this exercide, I will use simple binary cross-entropy loss functon

**Binary Cross-Entropy**


$$L_{BCE} = -{1 \over N} {\sum_{i=1}^N}({y_i*log(\hat{y}_i) + (1-y_i)*log(1-\hat{y}_i)})$$
$N$ is number of data point

$\hat{Y}$ is value predicted by model

$Y$ is value from real data point

```python
def bce(yhat, y):
    return - (y * np.log(yhat) + (1-y)*np.log(1-yhat)).sum() / len(y)

def derivative_bce(yhat, y, eps = 1e-6):
    return - (y / (yhat+eps) - (1-y) / (1-yhat+eps))
```

### Now, for training, we need a backpropagation pass

![alt text](images/backprop1.png "backpropagation")

![alt text](images/backprop.png "backpropagation")

```python
def backpropagation(X, y, lr):
    global W3, b3, W2, b2, W1, b1
    # Back prop
    N = len(X)
    
    z1, z2, z3, a1, a2, a3 = feedforward(X)
    
    dL_dYhat = derivative_bce(a3, y)
    da3_dz3 = derivative_sigmoid(z3)
    dL_dw3 = a2.T.dot(dL_dYhat * da3_dz3) 
    dL_db3 = (dL_dYhat * da3_dz3).sum()
    dL_da2 = (dL_dYhat * da3_dz3).dot(W3.T)
    
    da2_dz2 = derivative_tanh(z2)
    dL_dw2 = a1.T.dot(dL_da2 * da2_dz2)
    dL_db2 = (dL_da2 * da2_dz2).sum()
    dL_da1 = (dL_da2 * da2_dz2).dot(W2.T)
    
    da1_dz1 = derivative_tanh(z1)
    dL_dw1 = X.T.dot(dL_da1 * da1_dz1)
    dL_db2 = (dL_da1 * da1_dz1).sum()


    # Gradient descent for whole batch
    W3 += -lr * dL_dw3
    b3 += -lr * dL_db3
    W2 += -lr * dL_dw2
    b2 += -lr * dL_db2
    W1 += -lr * dL_dw1
    b1 += -lr * dL_db2
    
LR = 1e-4
# dummyX = np.random.randn(300, 2)
# dummyY = np.ones((300, 1))
# backpropagation(X, y, LR)
```

### Training the model

We will use gradient decent for optimize model

```python
def predict(X):
    # Forward prop
    input = X
    pre_act1 = X.dot(W1) + b1 # input layer -> hidden layer 1
    act1 = tanh(pre_act1) # activation function of hiddin layer 1
    pre_act2 = act1.dot(W2) + b2 # # hidden layer 1 -> hidden layer 2
    act2 = tanh(pre_act2) # activation function of hidden layer 2
    pre_act3 = act2.dot(W3) + b3 # hidden layer 2 -> ouput layer
    act3 = sigmoid(pre_act3) # activation function of output layer
    return act3
```

```python
num_iter = 5000
print_loss=True
# Train for each iteration
for i in range(0, num_iter):
    backpropagation(X, y, LR)
    # for j in range(len(X)):
    #     backpropagation(X[j:j+1, :], y[j:j+1, :], LR)
    yhat = predict(X)
    if print_loss and i % 500 == 1:
        print ("Loss after iteration %i: %f" % (i, bce(yhat, y)))
```

```python
def plot_decision_boundary(pred_func):
    # Set min and max values and give it some padding
    x_min, x_max = X[:, 0].min() - .5, X[:, 0].max() + .5
    y_min, y_max = X[:, 1].min() - .5, X[:, 1].max() + .5
    h = 0.01
    # Generate a grid of points with distance h between them
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))
    # Predict the function value for the whole gid
    Z = pred_func(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    # Plot the contour and training examples
    plt.contourf(xx, yy, Z, cmap=plt.cm.Spectral)
    plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.Spectral)

# Plot the decision boundary
plot_decision_boundary(lambda x: predict(x))
plt.title("Logistic Regression")
```
