### TODO 2: Create ImageDataGenerator

Read [this document](https://www.tensorflow.org/api_docs/python/tf/keras/preprocessing/image/ImageDataGenerator) and create an `ImageDataGenerator` object `train_datagen` for training with this property:
- `rescale: 1./255`
- `rotation_range: 20`
- `width_shift_range: 0.2`
- `height_shift_range: 0.2`
- `horizontal_flip: True`

For the `test_datagen` for testing, we only need `rescale=1./255`.

Questions:

```python
# Data Generator for training
train_datagen = tf.keras.preprocessing.image.ImageDataGenerator(
    rescale=1./255,
    rotation_range=20,
    width_shift_range=0.2,
    height_shift_range=0.2,
    horizontal_flip=True)

# Data Generator for evaluation
val_datagen = tf.keras.preprocessing.image.ImageDataGenerator(rescale=1./255)
```

Before create data flows from the data generators, we need to convert the `label` column in dataframe to string type because the `ImageDataGenerator` only works for string type

```python
train_data["label"] = train_data["label"].map(lambda x: str(x))
val_data["label"] = val_data["label"].map(lambda x: str(x))
```

### TODO 3: Create data flows

We will create 2 data flows for training and evaluation. In the same document of `ImageDataGenerator`, create the data flows `train_flow` and `val_flow` by the method `flow_from_dataframe`. You will consider those parameters: `dataframe`, `directory`, `x_col`, `y_col`, `class_mode`, `target_size`

For the `target_size`, it is typically `(224, 224)` in image classification task.

```python
# Data flow for training
train_flow = train_datagen.flow_from_dataframe(dataframe=train_data, 
                                        directory=data_dir + "/images", 
                                        x_col="image", 
                                        y_col="label",
                                        class_mode="categorical",
                                        target_size=(224,224))

# Data flow for evaluation
val_flow = val_datagen.flow_from_dataframe(dataframe=val_data, 
                                        directory=data_dir +"/images", 
                                        x_col="image", 
                                        y_col="label",
                                        class_mode="categorical",
                                        shuffle=False,
                                        target_size=(224,224))
```

### TODO 4: CNN Block

You need to define `ConvBlock` which is a convolutional block (conv + pooling layer). This block will be used multiple times in the network

References:
- [tf.keras.layers.Conv2D](https://www.tensorflow.org/api_docs/python/tf/keras/layers/Conv2D)
- [tf.keras.layers.MaxPool2D](https://www.tensorflow.org/api_docs/python/tf/keras/layers/MaxPool2D)
- [tf.keras.layers.ReLU](https://www.tensorflow.org/api_docs/python/tf/keras/layers/ReLU)

```python
class ConvBlock(tf.keras.Model):
    def __init__(self, filters, kernel, strides, padding='same'):
        '''
        Parameters
        ----------
        filters: int
            number of filter for convolutional layer
        kernel: int
            kernel size for convolutional layer
        strides: int
            stride of convolution layer
        padding: str
            padding of convolution layer
        
        '''
        
        super(ConvBlock, self).__init__()

        ### START CODE HERE
        
        # Create Conv2D layer
        self.cnn = tf.keras.layers.Conv2D(filters, kernel, strides, padding)
        
        # Create ReLU layer
        self.relu = tf.keras.layers.ReLU()

        # Create MaxPool2D layers with kernel size of 2, `same` padding
        self.pool = tf.keras.layers.MaxPool2D(2, padding='same')
        
        ### END CODE HERE
        
        
    def call(self, inputs):
        '''

        Forwarding the block
        
        Parameters
        ----------
        inputs: input tensor
        
        Returns
        -------
        tensor
            output tensor
        '''
        
        x = inputs

        ### START CODE HERE
        
        # Forward inputs to each layers
        
        x = self.cnn(x)
        x = self.relu(x)
        x = self.pool(x)

        ## END CODE HERE

        return x
```

### TODO 5: Define the whole network

We will use `ConvBlock` to define the whole network of CNN in the figure. The number of filters in each layer is up to you. For example, you can have 64, 128, 256, 512 for the 4 blocks.

For the flatten and Fully connected layer, you may want to check these references:
- [tf.keras.layers.Flatten](https://www.tensorflow.org/api_docs/python/tf/keras/layers/Flatten)
- [tf.keras.layers.Dense](https://www.tensorflow.org/api_docs/python/tf/keras/layers/Dense)

```python
class CNN(tf.keras.Model):
    def __init__(self, num_classes):
        
        super(CNN, self).__init__()
        
        ### START CODE HERE
        
        # Create convolution blocks
        self.block1 = ConvBlock(64, 3, 1, "same")
        self.block2 = ConvBlock(128, 3, 1, "same")
        self.block3 = ConvBlock(256, 3, 2, "same")
        self.block4 = ConvBlock(512, 3, 2, "same")
        
        # Flatten layer
        self.flatten = tf.keras.layers.Flatten()
        
        # Fully connected layer
        self.dense1 = tf.keras.layers.Dense(num_classes)
        
        self.blocks = [self.block1, self.block2, self.block3, self.block4, self.flatten, self.dense1]
        ### END CODE HERE

    def call(self, inputs):
        
        ### START CODE HERE
        
        # Forward inputs via each layer and assign to x
        x = inputs
        for b in self.blocks:
            x = b(x)
        
        ### END CODE HERE
        output = tf.nn.softmax(x)

        return output
```

##  Training

The code below is used to train the CNN model. We need to call the `fit_generator` to train the network. We can check the document of this method [here](https://www.geeksforgeeks.org/keras-fit-and-keras-fit_generator/)

```python
device = '/cpu:0' if len(tf.config.experimental.list_physical_devices('GPU')) == 0 else '/gpu:0'
batch_size = 64
train_flow.batch_size = batch_size
val_flow.batch_size = batch_size
steps_per_training = train_flow.n // batch_size
steps_per_evaluation = val_flow.n // batch_size
epochs = 10
num_classes = 6

with tf.device(device):

    # Create the model
    model = CNN(num_classes)
    
    # Create callback to save model with highest accuracy on the evaluation
    mcp = tf.keras.callbacks.ModelCheckpoint("my_model.h5", monitor="val_accuracy",
                      save_best_only=True, save_weights_only=True)

    # Compile model
    model.compile(optimizer=tf.keras.optimizers.Adam(1e-3), loss='categorical_crossentropy',
                  metrics=['accuracy'])
    
    # Training
    model.fit_generator(train_flow, 
                        steps_per_epoch=steps_per_training, 
                        epochs=epochs,
                        verbose=1, 
                        validation_data=val_flow,
                        validation_steps=steps_per_evaluation, 
                        callbacks=[mcp])
```

### Predict some random images on the test set

The following method supports reading and converting the image to the input format that can be fed to the CNN model

```python
def load_image(image_path):

    # Read image
    image = cv2.imread(image_path)

    # Resize image
    image = cv2.resize(image, (224, 224))

    # Convert BGR to RGB
    image = image[:, :, ::-1]

    # Normalize image
    image = image.astype('float32') * 1./255

    # Transform (224, 224, 3) to (1, 224, 224, 3)
    image = image[np.newaxis]
    return image
```

```python
# Dictionary ánh xạ nhãn và index
labels = train_flow.class_indices

# Ánh xạ indices về nhãn đúng
labels = dict((v,k) for k,v in labels.items())
```

## Transfer learning

There is a very detail tutorial about transfer learning from tensorflow. You can check it [here](https://www.tensorflow.org/tutorials/images/transfer_learning).

The sample code below will help you use the pretrained model as a feature extractor. You need to create the last Dense layer.

```python
def mobilenetv2():
    base_model = tf.keras.applications.MobileNetV2(input_shape=(224, 224, 3),
                                               include_top=False,
                                               weights='imagenet')
    
    # Using global average pooling
    x = base_model.output
    x = tf.keras.layers.GlobalAveragePooling2D()(x)

    # Adding last FC layer
    predictions = tf.keras.layers.Dense(11, activation='softmax')(x)

    # Combine them as the final model
    model = tf.keras.models.Model(base_model.input, predictions)
    return model
```

```python
model_mbnet = mobilenetv2()
print(model_mbnet.summary())
```

### TODO 6: Create model with EfficientNetB3
Implement the `efficientb3()` method below to create a model with EfficientNet-B3 as feature extractor

```python
def efficientnetb3():
    base_model = tf.keras.applications.efficientnet.EfficientNetB3(input_shape=(224, 224, 3),
                                               include_top=False,
                                               weights='imagenet')
    
    # Using global average pooling
    x = base_model.output
    x = tf.keras.layers.GlobalAveragePooling2D()(x)

    # Adding last FC layer
    predictions = tf.keras.layers.Dense(11, activation='softmax')(x)

    # Combine them as the final model
    model = tf.keras.models.Model(base_model.input, predictions)
    return model
```

### TODO 7: Create model with Resnet-50
Implement the `resnet50()` method below to create a model with ResNet-50 as feature extractor

```python
def resnet50():
    base_model = tf.keras.applications.resnet50.ResNet50(input_shape=(224, 224, 3),
                                               include_top=False,
                                               weights='imagenet')
    
    # Using global average pooling
    x = base_model.output
    x = tf.keras.layers.GlobalAveragePooling2D()(x)

    # Adding last FC layer
    predictions = tf.keras.layers.Dense(11, activation='softmax')(x)

    # Combine them as the final model
    model = tf.keras.models.Model(base_model.input, predictions)
    return model
```
